1.Which country has the most number of players

```python


most_players_country = data['Nationality'].value_counts().idxmax()
print("Country with the most players:", most_players_country)
```

2.Plot a bar chart of 5 top countries with the most number of players.

```python
top_countries =data['Nationality'].value_counts().head(6)
top_countries.plot(kind='bar')
plt.title('Top 5 Country with Most Players')
plt.xlabel("Country")
plt.ylabel("Number of Players")
plt.show()
```

3.Which player has the highest salary?

```python
data['Release Clause'] = data['Release Clause'].str.replace('€', '0000').str.replace('M', '0000000')
data['Release Clause']= pd.to_numeric(data['Release Clause'],errors='coerce')
```

```python
highest_salary_player =data.loc[data['Release Clause'].idxmax()]
print("Player with the highest Release Clause(salary):",
highest_salary_player['Name'])
print('Release Clause:',
highest_salary_player['Release Clause'])
```

4.Plot a histogram to get the salary range of the players. 

```python
data['Release Clause'].hist(bins=10)
plt.title("Salary Range of Players")
plt.xlabel("Salary")
plt.ylabel("Number of Players")
plt.show()
```

5.Who is the tallest player in the fifa? 

```python
print(data['Height'].head())
```

```python

def convert_height(height):
   if isinstance(height, str):
     feet, inches = height.split("'")
     return int(feet) * 30.48 + int(inches.replace('"', '')) * 2.54
     return float(height)

data['Height'] = data['Height'].apply(convert_height)
```

```python
tallest_player = data.loc[data['Height'].idxmax()]
print("Tallest player:", tallest_player['Name'])
```

6.Which club has the most number of players? 

```python
most_players_club = data['Club'].value_counts().idxmax()
print("Club with the most players:", most_players_club)
```
